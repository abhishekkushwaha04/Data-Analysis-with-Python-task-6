```python
#Lets find the answer of first question.

#1.How many people have heart disease and how many people doesn't have heart disease?

#Getting the Values
df.target.value_counts()
```

```python
# Question 5: What is the average age of people with heart disease?
avg_age_with_heart_disease = df[df['target'] == 1]['age'].mean()
print(f"The average age of people with heart disease is {avg_age_with_heart_disease:.2f} years.")
```

```python
#Plotting the Average Age of People with Heart Disease
plt.figure(figsize=(6, 4))
plt.bar(['Average Age'], [avg_age_with_heart_disease], color='salmon')
plt.ylabel('Age')
plt.title('Average Age of People with Heart Disease')
plt.show()
```

```python
# Question 6: What is the maximum heart rate achieved by people with and without heart disease?
max_heart_rate = df.groupby('target')['thalach'].max()
print("The maximum heart rate achieved by people with heart disease is:", max_heart_rate[1], "bpm")
print("The maximum heart rate achieved by people without heart disease is:", max_heart_rate[0], "bpm")
```

```python
# Question 6: Plotting the Maximum Heart Rate Achieved by People with and without Heart Disease
plt.figure(figsize=(6, 4))
max_heart_rate.plot(kind='bar', color=['teal', 'orange'])
plt.ylabel('Max Heart Rate (bpm)')
plt.xticks(ticks=[0, 1], labels=['Without Heart Disease', 'With Heart Disease'], rotation=0)
plt.title('Maximum Heart Rate by Heart Disease Status')
plt.show()
```

```python
heart_disease_count = df['target'].value_counts()
```
